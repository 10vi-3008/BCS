```python

numbers = list(map(int, input("Enter numbers separated by spaces: ").split()))

print("List of numbers:", numbers)


def find_median(numbers):
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)

    if n%2 == 1:
        return sorted_numbers[n//2]
    else:
        mid1,mid2 = sorted_numbers[n//2 -1],sorted_numbers[n//2]
        return(mid1+mid2)/2

print("The median of given numbers is: ", find_median(numbers))
```

```python
values = list(map(int, input("Enter numbers separated by spaces: ").split()))
n = len(values)
sum_values = 0
for i in range(n):
    sum_values += values[i]

avg = sum_values/n
print("Avg of the given numbers is: ",avg)
```
